```python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [10, 5]
plt.rcParams["figure.autolayout"] = False
```

```python
def logistic_feedback(init, r, steps=100):
    ys = [ init ]
    xs = [ 0 ]

    y = init

    for x in range(1, steps + 1):
        x = x

        # Add to x series
        xs.append(x)

        # Compute logistic output
        y = logistic_feedback_inner(y, r)
        ys.append(y)

    return (xs, ys)
```

```python
def logistic_feedback_inner(x, r):
    return r * x * (1 - x)
```

```python
x, y = logistic_feedback(0.4, 3.75, steps=50)
plt.grid()
plt.plot(x, y)
plt.show()
```
